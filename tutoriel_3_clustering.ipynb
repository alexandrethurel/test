## Qu'est-ce que le clustering ? <a class="anchor" id="chapter1"></a>

___

Le clustering est un des outils statistiques les plus utilis√©s, c'est un processus math√©matique ou informatique utilis√© pour diviser des donn√©es en diff√©rents groupes similaires. √Ä l'int√©rieur des groupes, les donn√©es seront regroup√©es selon une ou plusieurs caract√©ristiques communes.

Par exemple, dans la partie "Qu'est ce qu'une donn√©e ?", √† partir d'une base de donn√©es contenant quelques informations (poils, dents, nombre de pattes, etc...) sur des animaux, on dissocie clairement les mammif√®res, les poissons, les insectes ou les oiseaux, sans que cette information soit fournie.

La mesure de proximit√© entre deux donn√©es est g√©n√©ralement la distance euclidienne. La formule est la m√™me qu'au coll√®ge, √† la diff√©rence pr√®s qu'il n'y pas deux dimensions (x et y), mais qu'il y en a autant que de descripteurs (colonnes) dans notre jeu de donn√©es.

Le but de ce notebook est de vous guider dans la r√©alisation d'une m√©thode Clustering, sur un vrai jeu de donn√©es.

### Choisir son type de clustering <a class="anchor" id="section_3_1"></a>

Selon le public vis√© par vos r√©sultats, on peut utiliser diff√©rents types de clustering. Seules les 2 m√©thodes les plus connues seront abord√©es, ce ne sont pas les seules, vous pourrez retrouver une liste **[ici](https://www.analytixlabs.co.in/blog/types-of-clustering-algorithms/)**.


#### Les m√©thodes hi√©rarchiques. <a class="anchor" id="section_3_1_1"></a>

Afin de r√©aliser une classification hi√©rarchique, le programme va associer les deux informations les plus similaires en un groupe. Ensuite, la proximit√© entre chaque information et groupe sera calcul√©e. 

Il existera alors deux cas de figure :

- la proximit√© la plus importante est entre deux informations.
- la proximit√© la plus importante est entre un groupe et une information.

Dans le premier cas, les deux informations formeront un groupe, comme lors de la premi√®re √©tape, tandis que dans le second cas, l'information rejoindra le groupe similaire.

Cette op√©ration se r√©alisera en boucle, jusqu'√† ce qu'il ne reste plus que 2 groupes √† comparer.

Le r√©sultat graphique de ces m√©thodes est en dendrogramme  (ci-dessous). Les traits reliant les points signifient qu'ils appartiennent au m√™me groupe. La hauteur des traits repr√©sente une "mesure de la proximit√©", plus les traits sont hauts, moins les groupes, ou les informations, sont similaires. On peut aussi en d√©duire qu'ils ont √©t√© associ√©s plus tard par l'algorithme. 

Pour changer le nombre de clusters choisis, on peut le faire visuellement :

 - on peut choisir un nombre de clusters voulu (ici, 2) :

![title]( )

- ou choisir uniquement les groupes et les informations dont la proximit√© est minimum √† une valeur (ici, 5) :

![title]( )

Visuellement, le r√©sultat est plus efficace qu'avec les autres m√©thodes, puisqu'on peut changer le nombre de clusters (groupes) voulu sans avoir √† relancer le programme. Cependant, cela devient tr√®s co√ªteux en ressources et tr√®s peu visible lorsqu'on a trop de donn√©es. Les m√©thodes hi√©rarchiques sont donc des m√©thodes plus visuelles, √† privil√©gier lorsqu'on a un faible nombre de donn√©es.


####  Les m√©thodes centro√Ødes <a class="anchor" id="section_3_1_2"></a>

Ce sont les m√©thodes le plus utilis√©es.

Pour un clustering √† k clusters (le nombre de clusters √† choisir n'est pas al√©atoire, la suite du tutoriel vous pr√©sentera comment le choisir.) :

1. Le programme va placer k points al√©atoires appel√©s *centro√Ødes*, sur un graphique.
2. Les informations seront plac√©es dans le groupe du centro√Øde le plus proche.
3. Le programme va calculer la distance entre les donn√©es et la moyenne des donn√©es des groupes. 
4. Les donn√©es seront de nouveau attribu√©es au groupe dont la moyenne est la plus proche. 
5. Cette op√©ration sera enti√®rement r√©p√©t√©e, en commen√ßant avec de nouveaux points al√©atoires. 
6. L'op√©ration ayant fourni la distance moyenne la plus faible entre les informations sera choisie.

La qualit√© de nos r√©sultats augmente donc en fonction du nombre de fois que l'on va r√©p√©ter l'op√©ration.


<center><img src=" "/></center>

<center>figure 3 : m√©thodes centro√Ødes (larevueia.fr)</center>

C'est la m√©thode la plus rapide, des centaines de milliers de donn√©es pourront √™tre classifi√©es en quelques secondes. Cependant, obtenir un r√©sultat graphique des clusters est plus difficile, car cela n√©cessite d'utiliser une ACP (ou ses variantes), ce qui peut √™tre tr√®s impr√©cis. Les m√©thodes centro√Ødes sont donc optimales lorsqu'on a un tr√®s grand nombre de donn√©es et que le public vis√© √† des bases en statistiques.

### Selection des variables

```python
import pandas as pd

donnees = pd.read_csv('Data/master.csv', sep =';') # importation des donn√©es, le s√©parateur est ";"
donnees.head(5)                                    # v√©rification 
```

Afin de r√©aliser un clustering pertinent, il convient de choisir des variables toutes aussi pertinentes.

Ici, pour la selection des variables, nous nous appuierons sur : </br> 

- un indicateur de ch√¥mage **IDC**, qui d√©pendra du taux d'insertion, du pourcentage d'emploi stable et du taux de ch√¥mage observ√© dans la r√©gion. </br> C'est-√†-dire : taux d'insertion * pourcentage d'emploi stable / taux de ch√¥mage dans la r√©gion  </br></br>

- un indicateur de salaire **IDS**, qui d√©pendra du salaire net m√©dian des employ√©s √† temps plein et du salaire net m√©dian observ√© dans la r√©gion.</br> C'est-√†-dire : salaire net  m√©dian des employ√©s √† temps plein apr√®s 30 mois / salaire net m√©dian observ√© dans la r√©gion  


Les donn√©es sur les formations que nous utiliserons seront donc : le domaine, la discipline, l'identifiant (afin de nommer les formations), le taux d'insertion, le % d'emploi stable, le salaire net mensuel m√©dian  des emplois √† temps plein, le taux de ch√¥mage r√©gional, le salaire net mensuel m√©dian r√©gional (afin de faire nos indicateurs).

### Nettoyage des donn√©es

Parmi nos donn√©es, certaines sont issues d'√©l√®ves ayant eu leur master depuis 30 mois et d'autres depuis 18 mois, elles ne sont donc pas comparables !


```python
donnees[["situation"]].value_counts("situation") # On regarde combien de formations sont concern√©es par chacun des cas.
```

On voit qu'il y a plus de promotions d'√©l√®ves interrog√©s qui ont leur master depuis 30 mois (8 873) que de promotions d'√©l√®ves qui ont leur master depuis 18 mois (6 848). Cependant, on ne va prendre que les formations dans le 2√®me cas, car cela repr√©sente une situation √† plus court terme.


```python
donnees_court_terme = donnees[donnees["situation"] == "18 mois apr√®s le dipl√¥me"]
```

On va ensuite regarder si on a des valeurs manquantes.

```python
donnees_pertinentes = donnees_court_terme[["academie","domaine",
                                           "discipline","taux_dinsertion","emplois_stables",
                                           "salaire_net_median_des_emplois_a_temps_plein",
                                           "taux_de_chomage_regional","salaire_net_mensuel_median_regional"]]
donnees_pertinentes.isna().sum() # Nombre de donn√©es manquantes
```

Les donn√©es manquantes √©tant probl√©matiques, on va les supprimer. Il en va de m√™me pour toutes les fois on les valeurs non renseign√©es not√©es ".", "ns", ou "nd".

```python
import numpy as np

donnees_pertinentes2 = donnees_pertinentes.replace("ns",np.NaN,regex = True)
donnees_pertinentes3 = donnees_pertinentes2.replace("nd",np.NaN,regex = True)
donnees_pertinentes3["salaire_net_mensuel_median_regional"] = (donnees_pertinentes3["salaire_net_mensuel_median_regional"]
                                                               .str.replace(" ",""))
donnees_pertinentes4 = donnees_pertinentes3.replace("\\.",np.NaN,regex = True)



donnees_pertinentes_sansnan = donnees_pertinentes4.dropna()

donnees_pertinentes_sansnan.isna().sum() # v√©rification du nombre de donn√©es manquantes
```

Il n'y a plus de valeurs manquantes, on peut cr√©er nos indicateurs.


```python
donnees_pertinentes_sansnan = donnees_pertinentes_sansnan.copy(); 
donnees_pertinentes_sansnan["IDC"] = (donnees_pertinentes_sansnan["taux_dinsertion"].astype(float) *
                                      donnees_pertinentes_sansnan["emplois_stables"].astype(float) /
                                      donnees_pertinentes_sansnan["taux_de_chomage_regional"].astype(float)) / 1000
```

```python
donnees_pertinentes_sansnan = donnees_pertinentes_sansnan.copy(); 
donnees_pertinentes_sansnan["IDS"] = (donnees_pertinentes_sansnan["salaire_net_median_des_emplois_a_temps_plein"].astype(float)
                                      / donnees_pertinentes_sansnan["salaire_net_mensuel_median_regional"].astype(float))
```

Les donn√©es que nous allons utiliser pour faire le clustering seront donc les suivantes :

```python
donnees_clustering  = donnees_pertinentes_sansnan[['IDC',"IDS"]] # On ne selectionne que nos indicateurs
```

### R√©duire ses donn√©es <a class="anchor" id="section_4_1"></a>

Afin de mesurer la proximit√© entre les donn√©es, l'algorithme va s'appuyer sur la distance euclidienne entre les donn√©es. Cependant, les donn√©es ne sont pas du tout comparables, nos deux indicateurs ne sont pas comparables, car ils n'ont pas la m√™me √©chelle. C'est ici qu'intervient la r√©duction des donn√©es.


R√©duire ses donn√©es est primordial, cela consiste √† mettre toutes nos donn√©es quantitatives entre 0 et 1, afin de les mettre √† la m√™me √©chelle, tout en gardant la notion d'√©cart entre les valeurs.

Par exemple, avoir un 18/20 en histoire-g√©ographie peut sembler plus difficile (ou plus simple, pour certains !) au fait d'avoir un 18/20 en espagnol, il est donc difficile de les comparer. Cependant, gr√¢ce au principe de r√©duction et √† l'aide du nombre de notes, de la moyenne de la classe et de l'√©cart-type, on peut estimer combien un 18 en espagnol vaut en histoire-g√©ographie.


Sur Python, on r√©duit les donn√©es de cette mani√®re :


```python
from sklearn.preprocessing import MinMaxScaler                              # Importation du module n√©cessaire

minmax_scale = MinMaxScaler().fit(donnees_clustering) 
donnees_clustering_normalisees = minmax_scale.transform(donnees_clustering) # Application de la r√©duction
donnees_clustering_normalisees[0:5]                                         # Affichage des 5 premiers r√©sultats
```

## Choisir son nombre de clusters <a class="anchor" id="section_4_2"></a>

___

Le nombre de clusters va d√©terminer en combien de groupes vous allez s√©parer vos donn√©es. C'est un des param√®tres qui va le plus affecter la qualit√© de votre analyse, un nombre de clusters mal-choisi peut m√™me vous induire en erreur. 

Plus le nombre de clusters est √©lev√©, moins il y aura de donn√©es dans les groupes. Les r√©sultats seront alors plus pertinents, mais on risque de cr√©er des "mauvais" groupes. Imaginez demander 100 groupes pour diff√©rencier 100 personnes, il n'y aurait qu'une personne par groupes, ils seront donc inexploitables.

Plus le nombre de clusters est faible, plus il y aura de donn√©es dans le groupe. Les r√©sultats montreront donc des diff√©rences notables, au risque de donner des g√©n√©ralit√©s. 

Il existe plusieurs m√©thodes qui permettent d'obtenir des clusters repr√©sentatifs et une probabilit√© faible de "mauvais clusters". Deux d'entre elles vous seront pr√©sent√©es.


### M√©thode du coude ou *elbow method* <a class="anchor" id="section_4_2_1"></a>

Pour cette m√©thode, le programme va r√©aliser le clustering avec des nombres de clusters diff√©rents. Pour chaque clustering r√©alis√©s, il va calculer la distance entre les centres de chaque groupe ainsi que la distance entre chaque information et le centre de son groupe. Ces deux mesures sont appel√©es *distorsion* et *inertie*. Pas de soucis, derri√®re les calculs compliqu√©s se cachent une interpr√©tation tr√®s facile ! 


Le code pour r√©aliser une m√©thode du coude est le suivant :

```python
import matplotlib.pyplot as plt # libraries n√©cessaires
from sklearn.cluster import KMeans

wcss = {}
  
for k in range(2,100):
    model = KMeans(n_clusters=k)
    model.fit(donnees_clustering_normalisees)
    wcss[k] = model.inertia_
    
plt.plot(wcss.keys(), wcss.values(), 'gs-')
plt.xlabel('Values of "k"')
plt.ylabel('WCSS')
plt.show()
```

La m√©thode s'appelle "M√©thode du coude" car il faut voir le r√©sultat du graphique comme un bras. On va alors s√©lectionner la valeur de k au ¬´coude¬ª de ce bras, c‚Äôest-√†-dire le point apr√®s lequel la distorsion/inertie ne diminue plus beaucoup.

![title]( )


Ici, la valeur optimale serait entre 15 et 17, mais il est possible de prendre jusqu'√† 25 ans probl√®me.


### Score de silhouette  <a class="anchor" id="section_4_2_2"></a>

Comme la m√©thode du coude, elle va appliquer le clustering sur plusieurs nombres de clusters. Cependant, √† chaque √©tape, elle va regarder √† quel point l'information est similaire aux autres informations de son cluster, ensuite, elle va r√©aliser une moyenne de tous les points et donner un *score de silhouette*, qui devra √™tre le plus √©lev√© possible.


Le code pour obtenir le score de silhouette est le suivant :


```python
from sklearn.metrics import silhouette_score

plage_clusters = list(range(5, 100))
silhouette_avg = []
for k in plage_clusters:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(donnees_clustering_normalisees)
    cluster_labels = kmeans.labels_
    silhouette_avg.append(silhouette_score(donnees_clustering_normalisees, cluster_labels))

plt.plot(plage_clusters,silhouette_avg,"bx-")
plt.xlabel("Nombre de clusters") 
plt.ylabel("Score de silhouette") 
plt.title("M√©thode de la silhouette")
plt.show()
```

Selon cette m√©thode, le nombre optimal de clusters serait aux alentours de 40, les valeurs propos√©es par le r√©sultat des deux m√©thodes est diff√©rent mais coh√©rent. Afin de choisir entre les deux, r√©ferez-vous √† la diff√©rence entre un nombre de clusters √©lev√© par rapport √† un nombre de clusters faible.


> Important √† savoir : il peut √™tre n√©cessaire d'augmenter ou de r√©duire le nombre de clusters par rapport au "nombre optimal", qui n'est pas infaillible, bien que la proposition d'un nombre optimal de clusters trop faible est signe d'un manque de donn√©es ou de descripteurs (variables) pertinents. </br>

> Par exemple, si l'on observe l'√¢ge d'une population, un nombre de clusters faible peut dissocier les enfants des adultes. Un nombre de clusters consid√©r√© optimal peut s√©parer enfants, adultes et ain√©s. Un nombre de clusters tr√®s √©lev√© s√©parerait nourrissons, enfants, adolescents, adultes et ain√©s, ou donnerait des r√©sultats sans aucune pertinence.


Le choix du nombre de clusters est donc une des √©tapes les plus importantes, sur laquelle vous devez passer du temps et r√©aliser plusieurs essais.


Avec nos donn√©es, on va utiliser 15 clusters.


```python
import random # le clustering placant les premiers points al√©atoirement, on "fixe " l'al√©atoire afin de pouvoir reproduire

kmeans = KMeans(n_clusters = 15,n_init = 10000,random_state=27).fit(donnees_clustering_normalisees) # 15 clusters,
                                                                     # on relance l'op√©ration 10 000 fois
donnees_pertinentes_sansnan["cluster"] = kmeans.labels_
donnees_pertinentes_sansnan.head(5)
```

Apr√®s v√©rification, nos formations sont bien associ√©es √† un cluster.
